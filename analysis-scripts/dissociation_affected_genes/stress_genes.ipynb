```python
import pandas as pd
```

```python
pm_genes = pd.read_csv("./genes_affected_by_TPM.csv",skiprows=2)
pm_genes = pm_genes[~pm_genes["Gene"].isna()]

pm_genes
```

```python
dissociation_genes_vandenBrink = pd.read_excel("./genes_affected_by_dissociation.xlsx")
dissociation_genes_vandenBrink['Gene'] = [g.split("_")[0].upper() for g in dissociation_genes_vandenBrink['GeneName']]
dissociation_genes_vandenBrink = dissociation_genes_vandenBrink[~dissociation_genes_vandenBrink["Gene"].isna()]
dissociation_genes_vandenBrink
```

```python
dissociation_genes_vanVelthoven = pd.read_excel("./MuSCs_dissociation.xlsx",skiprows=1)
dissociation_genes_vanVelthoven = dissociation_genes_vanVelthoven[dissociation_genes_vanVelthoven["log2FC.sort.6h"]>0]
dissociation_genes_vanVelthoven = dissociation_genes_vanVelthoven[~dissociation_genes_vanVelthoven["Symbol"].isna()]
dissociation_genes_vanVelthoven["Symbol"] = dissociation_genes_vanVelthoven["Symbol"].str.upper()

dissociation_genes_vanVelthoven
```

```python
stress_genes = list(pm_genes["Gene"]) + \
    list(dissociation_genes_vandenBrink["Gene"]) + \
    list(dissociation_genes_vanVelthoven["Symbol"])
stress_genes = list(set(stress_genes))
stress_genes = [g for g in stress_genes if isinstance(g, str)]
stress_genes
```
