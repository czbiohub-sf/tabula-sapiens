# Load Pandas & Scanpy

```python
import pandas as pd
import scanpy as sc
```

# Read raw counts and clean it

```python
hpb = pd.read_csv("/mnt/ibm_lg/angela/sapiens/hPB/hPB.csv", index_col=0)
# remove extra htseq stuff
hpb = hpb[:-5]
hpb
```

Convert to gene symbols

```python
hpb.columns = ["sample_"+s.split("_A1_S111.homo.")[0] for s in hpb.columns]
```

```python
hpb.head()
```

# Get it as anndata

```python
hPB = sc.AnnData(hpb.transpose())
hPB
```

```python
hPB.obs.head()
```

```python
hPB.var.head()
```

```python
hPB.var['ensembl_id'] = hPB.var_names
hPB.var['ensembls'] = [g.split(".")[0] for g in hPB.var['ensembl_id']]
hPB.var
```

# Convert gene symbols

```python
import mygene
mg = mygene.MyGeneInfo()
```

```python
mygene_converter = mg.querymany(list(hPB.var['ensembls']),scopes='all', species='human', as_dataframe=True)
mygene_converter.loc[mygene_converter['notfound']==True,'symbol'] = mygene_converter.loc[mygene_converter['notfound']==True].index
```

```python
mygene_converter.tail()
```

```python
hPB.var = hPB.var.merge(
    mygene_converter.reset_index(),left_on='ensembls',right_on='query').sort_values(
    by='_score',ascending=False).drop_duplicates(
    'ensembl_id').set_index('symbol')
```

```python
hPB.var_names_make_unique()
```

# Load metadata

```python
metadata = pd.read_csv("/mnt/ibm_lg/angela/sapiens/hPB/RNA-seq-hPB.csv")
metadata
```

```python
samples_df = pd.DataFrame(hPB.obs_names,columns=['samples'])
```

```python
samples_df['matching_sample'] = [s.split(".homo.")[0].split("_")[0][:-1] for s in samples_df.samples]
samples_df['index'] = ["sample_"+s.split("_A1_S111.homo.")[0] for s in samples_df.samples]
samples_df.head()
```

# Write files to use with R

```python
hpb.merge(
    hPB.var.reset_index(),left_on='gene',right_on='ensembl_id').set_index('index').to_csv(
    "hPB_gene_count_table_with_gene_symbols.csv")
```

```python
samples_df.merge(metadata,on='matching_sample').set_index('index').to_csv("hPB_complete_metadata.csv")
```

# Run DGE

```python
hPB.obs = hPB.obs.join(samples_df.merge(metadata,on='matching_sample').set_index('index'))
```

```python
hPB.obs
```

```python
sc.tl.rank_genes_groups(hPB,groupby="Population", n_genes=1000)#, method = "t-test")
```

```python
result = hPB.uns['rank_genes_groups']
groups = result['names'].dtype.names

hPB_DGE = pd.DataFrame(
    {group + '_' + key: result[key][group]
    for group in groups for key in ['names','','logfoldchanges','pvals']})
hPB_DGE.to_csv("./hPB_rank_genes.csv")
hPB_DGE
```
